```python
import pandas as pd

def drop_duplicates(csv_file, columns):
  """
  Reads an excel file, drops duplicate rows based on given columns.

  Args:
    excel_file (str): Path to the excel file.
    columns (list): List of column names to consider for duplicate removal.

  Returns:
    pandas.DataFrame: DataFrame with duplicates removed.
  """

  df = pd.read_csv(csv_file)
  df_no_duplicates = df.drop_duplicates(subset=columns)
  return df_no_duplicates

# Example usage
file_path = '/content/drive/MyDrive/NASA Fellows Database Internship/nhfp grade data 2024.csv'  # Replace with actual file path
columns_to_check = ["Applicant ID", "PhD Institution", "Panel"]
result_df = drop_duplicates(file_path, columns_to_check)
result_df.to_csv('/content/drive/MyDrive/NASA Fellows Database Internship/2024 nhfp grades (no duplicates).csv', index=False)
```

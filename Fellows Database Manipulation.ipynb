```python
import pandas as pd

# Load the spreadsheets
spreadsheet1 = pd.read_excel('/content/drive/MyDrive/Updated List of Fellows (Maliyah Adams).xlsx')
spreadsheet2 = pd.read_excel('/content/drive/MyDrive/NASAPrizeFellowshipDemographicSurveyResponses2023apr.xlsx')

# Print column names to inspect them (just for confirmation, can be removed in final script)
# print("Spreadsheet1 columns:", spreadsheet1.columns)
# print("Spreadsheet2 columns:", spreadsheet2.columns)

spreadsheet1.columns = spreadsheet1.columns.str.strip()
spreadsheet2.columns = spreadsheet2.columns.str.strip()

# Helper function to extract first and last name
def extract_names(full_name):
    names = full_name.split()
    if len(names) == 1:
        return names[0], ''
    return names[0], names[-1]

# Extract first and last names from the name column in Spreadsheet2
spreadsheet2[['First Name', 'Last Name']] = spreadsheet2['What is your name?  (Leave blank if you prefer not to answer.)'].apply(lambda x: pd.Series(extract_names(str(x))))

# Columns from spreadsheet2 for merging
cols_to_merge = [
    'First Name', 'Last Name',
    'For Alumni: How would you characterize your current career path? (You may check multiple boxes.)',
    'What was your NASA prize fellowship host institution?  (If you changed hosts, just list the initial host.  Applicants leave BLANK.)',
    'What year was your Ph.D awarded?  (YYYY)'
]

# Merge based on First and Last names
merged_df = pd.merge(
    spreadsheet1,
    spreadsheet2[cols_to_merge],
    how='left',
    left_on=['First Name', 'Last Name'],
    right_on=['First Name', 'Last Name']
)

# Fill columns E, F, and G in Spreadsheet1 with data from Spreadsheet2
spreadsheet1['Work Email (.edu)'] = merged_df['For Alumni: How would you characterize your current career path? (You may check multiple boxes.)']
spreadsheet1['Personal Email (.com)'] = merged_df['What was your NASA prize fellowship host institution?  (If you changed hosts, just list the initial host.  Applicants leave BLANK.)']
spreadsheet1['Fellowship Year'] = merged_df['What year was your Ph.D awarded?  (YYYY)']

# Save the updated spreadsheet1
spreadsheet1.to_excel('/content/drive/MyDrive/updated_spreadsheet1.xlsx', index=False)
```

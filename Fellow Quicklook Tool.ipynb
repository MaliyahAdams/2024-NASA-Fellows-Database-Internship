**The code below prints Felllows from years 2019-2016 (descending order by year)**

```python
data2 = pd.read_excel(fellow_list, usecols='C,D,G,H,I,N,O' )
years = [2019, 2018, 2017, 2016]
data2 = data2[data2.iloc[:, 2].isin(years)]
#print(data2)
```

**The code below filters the Fellows from the years 2019 to 2016 who havent been categorized as a professor. This is the filtered data we will run through the Reviewer Extractor code below.**

**I'm going to turn this portion of the excel sheet into a csv to be used by the extractor code. There is a change implementd here to modify the .csv Name column to be in the format that is utilized by the extractor code**

```python
data2_filtered.to_csv('/content/drive/My Drive/filtered_fellows.csv', index=False)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

file_path = '/content/drive/My Drive/filtered_fellows.csv'
filtered_fellows = pd.read_csv(file_path)

filtered_fellows['Name'] = filtered_fellows['Last Name'] + ', ' + filtered_fellows['First Name']

filtered_fellows = filtered_fellows.drop(columns=['First Name', 'Last Name'])

cols = ['Name'] + [col for col in filtered_fellows.columns if col != 'Name']
filtered_fellows = filtered_fellows[cols]

filtered_fellows.to_csv(file_path, index=False)

# Print the first few rows to verify the changes
# print(df.head())
#print(filtered_fellows)

```
